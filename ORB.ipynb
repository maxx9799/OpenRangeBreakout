```python
buy = False
sell = False
buyorder = 0
exittime = "15:25"
nottradetime = "15:25"
stoplossbuy = 0
stoplosssell = 0
entrytime = "09:25"
entrytime1 = "09:15"
entrytime2 = "09:20"
entrytime3 = "09:25"
longentry = 0
shortentry = 0
stoplosshit = 0
cumprofit = 0
count = 0
```

```python
time = df["Time"]
```

```python
for i in range(len(time)):
    if (entrytime in df["Time"][i]):
        print("Time :", df["Time"][i], "on index", i)
        lowtemp = []
        lowtemp.append(df["Low"][i])
        lowtemp.append(df["Low"][i-1])
        lowtemp.append(df["Low"][i-2])
        print(lowtemp)
        stoplossbuy = min(lowtemp)
        
        #highTemp
        hightemp = []
        hightemp.append(df["High"][i])
        hightemp.append(df["High"][i-1])
        hightemp.append(df["High"][i-2])
        stoplosssell = max(hightemp)
        print(hightemp)
        
        print(" Low(stoplossbuy) :",str(stoplossbuy),"\n","High(stoplosssell) :",str(stoplosssell))
        print("-"*50)
        print()
        
    elif (df["High"].values[i]>stoplosssell and buy==False and buyorder==0 and entrytime1 not in df["Time"][i] 
          and entrytime2 not in df["Time"][i]and entrytime3 not in df["Time"][i] and nottradetime in df["Time"][i]):
        print("Order for Buy is placed at ", df["Time"][i],"on Date", df["Date"][i])
        buy = True
        buyorder=1
        count+=1


    elif (df["Low"].values[i]>stoplossbuy and sell==False and buyorder==0 and entrytime1 not in df["Time"][i] and entrytime2 not in df["Time"][i]and entrytime3 not in df["Time"][i] and nottradetime in df["Time"][i]):
        print("Order for Buy is placed at ", df["Time"][i],"on Date", df["Date"][i])
        sell = True
        buyorder=1
        count+=1
        
    elif (exittime in df["Time"][i]):
        if (buy == True and stoplosshit==0):
            print("Sell at exit time : ", df.Time[i])
            buy == False
            buyorder = 0
            cumprofit =  cumprofit+df["Close"][i] - stoplosssell
            print("cumm_profit : ", str(cumprofit))
            stoplosshit == 0
            
        elif (sell == True and stoplosshit==0):
            print("Sell at exit time : ", df.Time[i])
            buy == False
            buyorder = 0
            cumprofit =  cumprofit + stoplossbuy- df["Close"][i] 
            print("cumm_profit : ", str(cumprofit))
            stoplosshit == 0
            
    elif(sell==True and df["High"][i] > stoplosssell and stoplosshit==0):
        print("stoploss hit at", df["Time"][i])
        stoplosshit=1
        
    elif(sell==True and df["Low"][i] > stoplossbuy and stoplosshit==0):
        print("stoploss hit at", df["Time"][i])
        stoplosshit=1
        
    elif(stoplosshit==1 and buy==True):
        buy=False
        buyorder=0
        stoplosshit=0
        cumprofit=cumprofit-(stoplosssell-stoplossbuy)
        print("Cum_Profit", str(cumprofit))
        print("Time: ", df["Time"][i])
        
        
    elif(stoplosshit==1 and sell==True):
        buy=False
        buyorder=0
        stoplosshit=0
        cumprofit=cumprofit-(stoplosssell-stoplossbuy)
        print("Cum_Profit", str(cumprofit))
        print("Time: ", df["Time"][i])
    
print("cumprofit in the end ", str(round(cumprofit,2)))
print("No of trades taken in the period ", str(count))
```
